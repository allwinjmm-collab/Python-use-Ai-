```python
import cv2

# Open the default webcam (usually device 0)
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

print("Press 'q' to quit.")

while True:
    # Read frame-by-frame
    ret, frame = cap.read()
    
    if not ret:
        print("Error: Failed to capture image.")
        break

    # Display the resulting frame
    cv2.imshow('Webcam Feed', frame)

    # Exit loop when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```

```python
from skimage import io, color, filters, util
import matplotlib.pyplot as plt

# Step 1: Load the image
image = io.imread(r"C:/Users/Admin/Downloads/vijay.jpg")

# Step 2: Remove alpha channel if it exists
if image.shape[-1] == 4:
    image = image[:, :, :3]

# Step 3: Convert to grayscale
gray = color.rgb2gray(image)

# Step 4: Detect edges using Sobel filter
edges = filters.sobel(gray)

# Step 5: Invert the image (makes it look like a sketch)
sketch = util.invert(edges)

# Step 6: Show original and sketch side by side
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(image)
ax[0].set_title("Original Image")
ax[0].axis('off')

ax[1].imshow(sketch, cmap='gray')
ax[1].set_title("Pencil Sketch Effect")
ax[1].axis('off')

plt.show()

```

```python
import cv2
import matplotlib.pyplot as plt

# Load image in grayscale
image = cv2.imread(r"C:/Users/Admin/Downloads/vijay.jpg", cv2.IMREAD_GRAYSCALE)  # Replace with your image path

# Apply binary thresholding
_, thresh = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# Find contours
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on a copy of the original (converted to BGR for color drawing)
image_with_contours = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
cv2.drawContours(image_with_contours, contours, -1, (0, 255, 0), 2)

# Show results
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Thresholded Image")
plt.imshow(thresh, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Contours Detected")
plt.imshow(cv2.cvtColor(image_with_contours, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load original image
image = cv2.imread(r"C:/Users/Admin/Downloads/vijay.jpg",)  # Replace with your image path
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Step 1: Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Step 2: Apply median blur
gray_blur = cv2.medianBlur(gray, 7)

# Step 3: Detect edges using adaptive thresholding
edges = cv2.adaptiveThreshold(
    gray_blur, 255,
    cv2.ADAPTIVE_THRESH_MEAN_C,
    cv2.THRESH_BINARY,
    blockSize=9,
    C=2
)

# Step 4: Apply bilateral filter to the original image for a smooth effect
color = cv2.bilateralFilter(image, d=9, sigmaColor=250, sigmaSpace=250)

# Step 5: Combine edges and color
cartoon = cv2.bitwise_and(color, color, mask=edges)

# Convert to RGB for display
cartoon_rgb = cv2.cvtColor(cartoon, cv2.COLOR_BGR2RGB)

# Show result
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Cartoonified Image")
plt.imshow(cartoon_rgb)
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv2.imread(r"C:/Users/Admin/Downloads/vijay.jpg")  # Replace with your image path
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert to HSV color space
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define green screen color range in HSV
lower_green = np.array([35, 40, 40])
upper_green = np.array([85, 255, 255])

# Create mask for green areas
mask = cv2.inRange(hsv, lower_green, upper_green)

# Invert mask to get the foreground
mask_inv = cv2.bitwise_not(mask)

# Extract foreground (person or object)
foreground = cv2.bitwise_and(image_rgb, image_rgb, mask=mask_inv)

# Create a plain white background (or any background)
white_bg = np.full(image.shape, 255, dtype=np.uint8)
white_bg_rgb = cv2.cvtColor(white_bg, cv2.COLOR_BGR2RGB)

# Combine background and foreground
background = cv2.bitwise_and(white_bg_rgb, white_bg_rgb, mask=mask)
final_result = cv2.add(foreground, background)

# Show result
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.title("Original")
plt.imshow(image_rgb)
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Mask (Green Screen)")
plt.imshow(mask, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Background Removed")
plt.imshow(final_result)
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np

def cartoonify(image_path, output_path):
    # Read the image
    img = cv2.imread(r"C:/Users/Admin/Downloads/vijay.jpg")

    if img is None:
        print("Could not read image.")
        return

    # Resize for faster processing
    img = cv2.resize(img, (640, 480))

    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Apply median blur
    gray_blur = cv2.medianBlur(gray, 7)

    # Detect edges using adaptive thresholding
    edges = cv2.adaptiveThreshold(gray_blur, 255,
                                  cv2.ADAPTIVE_THRESH_MEAN_C,
                                  cv2.THRESH_BINARY,
                                  blockSize=9,
                                  C=2)

    # Apply bilateral filter to reduce color palette but keep edges sharp
    color = cv2.bilateralFilter(img, d=9, sigmaColor=300, sigmaSpace=300)

    # Combine edges and color image
    cartoon = cv2.bitwise_and(color, color, mask=edges)

    # Save output
    cv2.imwrite(output_path, cartoon)
    print(f"Cartoonified image saved as {output_path}")

# Usage
cartoonify('input.jpg', 'cartoon_output.jpg')
```
